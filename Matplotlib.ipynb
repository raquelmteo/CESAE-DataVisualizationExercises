Show a bar chart of the popularity of the following programming languages. <br>
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'] <br>
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7] <br>
x_pos = [i for i, _ in enumerate(x)] <br>

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, popularity, color="b")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages Worldwide \nOct 2017 compared to the previous year")
plt.grid(True, alpha=0.7, color='red')
plt.minorticks_on()
plt.grid(True, which='minor', linestyle='dotted', alpha=0.5, color="black")

plt.xticks(x_pos, x)
plt.xlim(-0.40, len(x) + 0.4)
plt.ylim(0, max(popularity) + 0.4)
```

Create a score bar chart by group and gender.<br>
n_groups = 5 <br>
men_means = (22, 30, 33, 30, 26)<br>
women_means = (25, 32, 30, 35, 29)<br>

```python
n_groups = 5 
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

bar_width = 0.35
index = np.arange(n_groups)

plt.bar(index, men_means, bar_width, label='Homens', color='green')
plt.bar(index + bar_width, women_means, bar_width, label='Mulheres', color='red')

plt.xlabel('Pessoa')
plt.ylabel('Pontua√ß√£o por pessoa')
plt.legend(loc='upper right')
plt.xlim(-0.19, 5.8)
plt.xticks(index + bar_width/2, ('G1', 'G2', 'G3', 'G4', 'G5'))
```

Create a pie chart of the gold medal achievements of the five most successful countries at the 2016 Olympic Games. Read the data from an ex6.csv file.

```python
labels = ['United States', 'Great Britain', 'China', 'Russia', 'Germany']
values = [34.1, 20, 19.3, 14.1, 12.6]
colors = ['#1F77B4', "#FF7F0F", "#2BA02B", "#D62727", "#8C564C"]

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=120, shadow=True, colors=colors, explode = (0.1, 0, 0, 0, 0))

plt.title('Top 5 countries gold medal\nat the 2016 Summer Olympics')
```
