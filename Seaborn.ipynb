Let's work with a famous titanic dataset for these exercises:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

Recreate the graphics below using the titanic dataframe.


```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
sns.histplot(data = titanic, x = 'fare', color = 'red', binwidth=20)
```

```python
sns.boxplot(x = "class", y = "age", data = titanic, hue = "class", legend = False, palette = 'rainbow', flierprops = {"marker": "D", "markersize": 4, "markerfacecolor": "black"})
```

```python
sns.swarmplot(data=titanic, x="class", y="age", hue="class", legend=False, palette= "Set2", size = 4)
```

```python
sns.countplot(x = 'sex', data = titanic, hue = 'sex', legend = False)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
```
